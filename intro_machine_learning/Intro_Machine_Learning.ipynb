# Scikit-Learn 



```python

# Aprendizado supervisionado

# importar dados do scikit-learn
from sklearn.datasets import load_boston

# Carregar dados
boston = load_boston()

# imprimir descrição dos dados
print(boston.DESCR)
```

```python
# imprimir primeiro conjunto de features contido no conjunto
print(boston.data[0])

# imprimir valor da primeira casa
print(boston.target[0])

# Usar apenas uma feature. As features estão contidas em linhas. Aqui faz-se um slice para usar apenas
# a sexta feature. Carrega-se esses dados na variável X.
X = boston.data[:,5]

# Preço. Carrega-se o preço das casas na variável y.
y = boston.target
```

```python
# importar matplotlib. Módulo de plotagem.
import matplotlib.pyplot as plt

# Faz com que o plot fique dentro da célula
%matplotlib inline

# Plotar a relação entre o preço das casas e o número médio de quartos na vizinhança.

# Definir tamanho da figura
plt.figure(figsize=(10,10))

# Scatter plota um gráfico de dispersão
plt.scatter(X, y) 

# Legenda dos eixos e título.
plt.xlabel("Media de quartos na vizinhanca")
plt.ylabel("Preco")
plt.title("Relacao entre preco e numero de quartos")

# Plotagem
plt.show()
```

# Modelos Paramétricos  X Modelos Não-Paramétricos

**Paramétricos**: representam o modelo como um conjunto de parâmetros. Podem ser representados por uma equação.
Ex: f(x1, x2, x3) = ax1 + bx2 + cx3 onde a, b e c são os parâmetros.

Vantagem: Leve (na maioria dos casos) porque uma vez treinado, não precisa dos dados.

Desvantagem: Tempo de treinamento pode ser alto.

**Não-Paramétricos**: o modelo é representado atrvés dos dados. Não possuem equação. Ex: KNN (k-nearest neighbors)

Vantagen: Rápido! (Não precisa de treinamento)

Desvantagem: Pesado (precisa armazenar os dados) 

# Paramétrico (Regressão Linear)

```python
# importa classe de regressão linear do scikit-learn
from sklearn.linear_model import LinearRegression

# import função de separação de dados
from sklearn.model_selection import train_test_split

# Reshaping (Ignorar)
X = X.reshape(-1, 1)

# Dividir dados em treinamento e teste. 70% dos dados ficam para treinamento e 30% para teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python

# Criar algoritmo. Instanciar classe LinearRegression
reg = LinearRegression()

# Treinar objeto
reg.fit(X_train, y_train)

# Fazer previsão (dados de treinamento)
y_pred = reg.predict(X_train)

```

```python
# Definir tamanho da figura
plt.figure(figsize=(10,10))

# Gráfico de dispersão
plt.scatter(X, y)

# Plotar modelo treinado
plt.plot(X_train, y_pred, color='red')

# Legendas
plt.xlabel("Media de quartos na vizinhanca")
plt.ylabel("Preco")
plt.title("Relacao entre preco e numero de quartos")
 
# Plotar
plt.show()
```

```python
# Avaliando modelo. Métrica é o R2 que tem valor máximo igual a 1. Quanto maior, melhor!!!
# Usar método score(x,y)

R2 = reg.score(X_test, y_test)

print("O R2 é {}".format(R2))
```

```python
# Parâmetros

a = reg.coef_

b = reg.intercept_

print("y = {}x + {}".format(a[0],b))
```

```python
# Usar todas as features
X = boston.data

# Dividir dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)

# Criar modelo
reg = LinearRegression()

# Treinamento
reg.fit(X_train, y_train)

# Fazendo previsões
y_pred = reg.predict(X_train)

# Avaliando modelo

R2 = reg.score(X_test, y_test)

print("O R2 é {}".format(R2))
```

# Não-Paramétrico (KNN)

```python
# Usar apenas uma feature. As features estão contidas em linhas. Aqui faz-se um slice para usar apenas
# a sexta feature. Carrega-se esses dados na variável X.
X = boston.data[:,5]

# Preço. Carrega-se o preço das casas na variável y.
y = boston.target
```

```python
# importa classe de knn do scikit-learn
from sklearn.neighbors import KNeighborsRegressor

# import função de separação de dados
from sklearn.model_selection import train_test_split

# Reshaping (Ignorar)
X = X.reshape(-1, 1)

# Dividir dados em treinamento e teste. 70% dos dados ficam para treinamento e 30% para teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
# Criar algoritmo. Instanciar classe LinearRegression
knn = KNeighborsRegressor(n_neighbors=3, weights='distance')

# Treinar objeto
knn.fit(X_train, y_train)

# Fazer previsão (dados de treinamento)
y_pred = knn.predict(X_train)

```

```python
# Definir tamanho da figura
plt.figure(figsize=(10,10))

# Gráfico de dispersão
plt.scatter(X, y)

# Plotar modelo treinado
plt.scatter(X_train, y_pred, color='red')

# Legendas
plt.xlabel("Media de quartos na vizinhanca")
plt.ylabel("Preco")
plt.title("Relacao entre preco e numero de quartos")
 
# Plotar
plt.show()
```

```python
# Avaliando modelo. Métrica é o R2 que tem valor máximo igual a 1. Quanto maior, melhor!!!
# Usar método score(x,y)

R2 = knn.score(X_test, y_test)

print("O R2 é {}".format(R2))
```

```python
# Usar todas as features
X = boston.data

# Dividir dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)

# Criar modelo
knn = KNeighborsRegressor(n_neighbors=3, weights='distance')

# Treinamento
knn.fit(X_train, y_train)

# Fazendo previsões
y_pred = knn.predict(X_train)

# Avaliando modelo

R2 = knn.score(X_test, y_test)

print("O R2 é {}".format(R2))
```
